```python
import pandas as pd
```

```python
temp = pd.read_csv("/content/drive/MyDrive/최종프로젝트/AWS_연평균기온풍속.csv", encoding='cp949')
temp
```

```python
temp = temp.drop('평균 풍속(m/s)', axis=1)
```

```python
temp = temp.groupby(['지점','지점명']).mean().reset_index().drop('일시', axis=1)
temp
```

```python
stnid = pd.read_csv('/content/drive/MyDrive/최종프로젝트/AWS_관측지점정보.csv', encoding='cp949')
stnid = stnid[['지점', '위도', '경도']]
stnid
```

```python
stnid = stnid.drop_duplicates(['지점'],keep='first')
```

```python
final = pd.merge(left=temp, right=stnid, how='inner', on='지점')
```

```python
final.to_csv('/content/drive/MyDrive/최종프로젝트/평균기온_위경도.csv', encoding='utf-8-sig')
```
