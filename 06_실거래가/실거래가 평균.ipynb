## 전체 데이터에 적용

```python
price = pd.read_csv('실거래가_raw.csv')
```

```python
price
```

```python
def make_mean(df):
    df = df[['Source.Name','지목','용도지역','거래금액(만원)']]
    df = df[df.isin({'지목' : ['전','답','임야']})['지목']]
    df = df[~df.isin({'용도지역' : ['자연녹지지역','개발제한구역','농림지역','용도미지정','보전관리지역']})['용도지역']]
    df['Source.Name'] = df['Source.Name'].apply(lambda x : x[:-4])
    df.rename(columns={'Source.Name':'시군'}, inplace=True)
    df = df.groupby('시군').mean()
    df = df.reset_index()
    return df
```

```python
price_mean = make_mean(price)
price_mean
```

```python
price_mean.to_csv('실거래가_평균.csv', encoding='utf-8-sig')
```
